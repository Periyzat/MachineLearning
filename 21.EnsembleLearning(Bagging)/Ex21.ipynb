```python
# now we will replace some text values above with number values 
df3 = df2.copy()
df3.RestingECG.replace(
{
    'Normal': 1,
    'ST': 2,
    'LVH': 3
}, 
inplace = True
)

df3.ST_Slope.replace(
{
    'Up': 1,
    'Flat': 2,
    'Down': 3
}, 
inplace = True)

df3.head()
```

```python
# for other values such as "Sex","ExerciseAngina" and "ChestPainType" we will use dummies encoding
df4 = pd.get_dummies(df3, drop_first=True)
df4.head()
```

```python
x=df4.drop(['HeartDisease'], axis='columns')
y=df4.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=20)
```

```python
# first tested it in SVC in cross_val_score
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
scores = cross_val_score(SVC(), x,y, cv=5) # setting k=5
scores.mean()
```

```python
# and then in BaggingClassifier 
# here it doesn't show many difference in result
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    base_estimator=SVC(),
    n_estimators=100,  # number of samples
    max_samples=0.8,   # 80% of all total x_train is used 
    oob_score = True,    # out of bag - which is 20% of x_train left
    random_state = 0
)

scores = cross_val_score(bag_model, x,y, cv = 5)
scores.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
scores = cross_val_score(DecisionTreeClassifier(), x,y, cv=5) # setting k=5
scores.mean()
```

```python
# But in DecisionTreeClassifier it shows big difference from 0.72% to 0.81
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=100,  # number of samples
    max_samples=0.8,   # 80% of all total x_train is used 
    oob_score = True,    # out of bag - which is 20% of x_train left
    random_state = 0
)

scores = cross_val_score(bag_model, x,y, cv = 5)
scores.mean()
```

```python
# But RandomForestClassifier itself shows high accuracy
from sklearn.ensemble import RandomForestClassifier
scores = cross_val_score(RandomForestClassifier(), x,y, cv=5) # setting k=5
scores.mean()
```
